# Reshaping DataFrame

```python
df_wide = pd.DataFrame({
    'Name': ['Angel', 'Gehendra', 'Kaushal'],
    'Math': [75, 90, 80],
    'Science': [90, 95, 100],
    'English': [93, 92, 97
                ],
})

print(df_wide)
```

```python
df_long = pd.melt(frame=df_wide,
                  id_vars=['Name'],
                  var_name='Subject',
                  value_name='Marks')

print(df_long)
```

## Pivot

```python
df_pivot = df_long.pivot(
    index = 'Name',
    columns=['Subject'],
    values="Marks"
)

df_pivot
```

```python
stacked_df = df_long.stack()
stacked_df
```

```python
unstacked_df = stacked_df.unstack()
unstacked_df
```

```python
pivot_table_df = pd.pivot_table(
    data=df_long,
    index = 'Name',
    columns=['Subject'],
    values="Marks",
    aggfunc='mean'
)

pivot_table_df
```

## Cateogrical Data

```python
df = pd.DataFrame({
    'EID': [101, 102, 103, 104],
    "Name": ["Ram", "Shyam", "Hari", "Sita"],
    "Level": ["Junior", "Senior", "Mid", "Junior"]
})

df.dtypes
```

```python
df['Level'] = df['Level'].astype("category")
df.dtypes
```

```python
df['Level'] = df["Level"].cat.rename_categories({
    "Junior":"Jr",
    "Senior":"Sr"
})

df.head()
```

```python
df['Level_code'] = df["Level"].cat.codes
print(df.head())
```

```python
encoded = pd.get_dummies(df['Level'], prefix='Lev')
df = pd.concat(objs=[df, encoded], axis=1)
print(df)
```
