```python
class CustomCounter:
    def __init__(self, limit):
        self.limit = limit
        self.current = 0
    
    def __iter__(self):
        return self

    def __next__(self):
        if self.current < self.limit:
            self.current +=  1
            return self.current
        else:
            raise StopIteration

obj = CustomCounter(5)
for num in obj:
    print(num, end=" ")
```

```python
def cubes(n):
    for i in range(n):
        yield n**3

print(cubes(5))
```
