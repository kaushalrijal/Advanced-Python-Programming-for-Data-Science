```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("tips.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
sns.boxplot(x="gender", y="total_bill", data=df, hue="gender")

plt.title("Total bill by gender")
plt.show()
```

```python
# pari plot
sns.pairplot(df, hue='gender', diag_kind="kde");
plt.show()
```

```python
# Heatmap

data = pd.DataFrame(
    [[10, 20, 30],
     [20, 15, 25],
     [30, 25, 5]],
     columns=['A', 'B', 'C'],
     index=['X', 'Y', 'Z']
)

# create heatmap
sns.heatmap(data, annot=True, cmap='gray')

plt.title("Simple Heatmap Example")
plt.show()
```
