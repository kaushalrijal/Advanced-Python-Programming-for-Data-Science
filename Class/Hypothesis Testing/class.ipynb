# Chi Square
$\chi^2$

```python
import numpy as np

#observed counts
observed = np.array([
    [250,200],
    [50, 1000]
])

# expected counts
expected = np.array([
    [90, 360],
    [210, 840]
])

# manually calculate chi-square

chi_square = ((observed - expected) ** 2 / expected).sum()
print(f"Chi-square value: {chi_square:.4f}")
```

```python
from scipy.stats import chi2_contingency

chi2, p, dof, exp = chi2_contingency(observed, correction=False)
# chi2, p, dof, exp = chi2_contingency(observed)

print(f"Chi-square: {chi2:.4f}")
print(f"p-value: {p:.4f}")
print(f"Degrees of freedom: {dof:.4f}")
print(f"Expected Values: {exp}")

if p < 0.05:
    print("Reject H0")
else:
    print("Failed to reject H0")
```

```python
from scipy.stats import chisquare

observed = [16, 20, 25, 14, 29, 28]
expected = [22] * 6

chi2_stat, p_value = chisquare(f_obs=observed, f_exp=expected)

print("Chi-square:", chi2_stat)
print("P-value:", p_value)

if p_value < 0.05:
    print("Reject H0: The die is biased.")
else:
    print("Fail to reject H0: The die is likely fair.")
```
