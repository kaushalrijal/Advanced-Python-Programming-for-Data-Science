```python
df = df.drop(columns=['Cabin'])
df
```

```python
df['Age'] = df['Age'].fillna(df['Age'].median())
```

```python
# df['Fare'] = df['Fare'].convert_dtypes(convert_floating=True)
df['Fare'] = pd.to_numeric(arg=df['Fare'], downcast='float')
```

```python
df['Fare'] = df['Fare'].fillna(df['Fare'].mean())
```

```python
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])
```

```python
plt.bar(x=df['Survived'].unique(), height=df['Survived'].value_counts(), color=['green', 'red'])
plt.show()
```

```python
sns.histplot(df['Age'], kde=True, bins=20)
plt.title("Bar Plot for Age Count")
plt.show()
```

```python
sns.boxplot(x="Gender", y="Fare", data=df)
plt.title("Fare by Gender")
plt.show()
```

```python
# scatterplot
sns.scatterplot(x="Age", y="Fare", hue="Survived", data=df)
plt.title("Age vs Fare")
plt.show()
```

```python
# correlation matrix
corr_mat = df.select_dtypes(include="number").corr()
sns.heatmap(corr_mat, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```
