```python
import pandas as pd
import csv
```

```python
def squares(x):
    for i in range(x):
        yield i**3

with open(file="large.csv", mode="w") as f:
    writer = csv.writer(f)
    writer.writerow(["Name", "Num", "Address"])
    for x in squares(100000000):
        writer.writerow(["Pari", x, "Taltira"])
        
```

```python
first_chunk = False
for chunk in pd.read_csv("large.csv", chunksize=10000):
    chunk["Salary"] = chunk['Num'].apply(lambda x: "50000" if x%2==0 else "23674")
    chunk.to_csv(
        "handled.csv",
        mode="w" if first_chunk else "a",
        header = first_chunk,
        index=False
    )

    first_chunk = True
```
