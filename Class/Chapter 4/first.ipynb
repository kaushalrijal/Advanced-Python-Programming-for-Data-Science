# Correlation
- Correlation is a standarized covariance
- Measures strength and direction of relationship
- Scale-free measure
>   +1 -> Perfect Positive Relationship

>   -1 -> Perfect Negative Relationship

>    0 -> No Linear Relationship

```python
import numpy as np
import pandas as pd

x = np.array([10,20,30,40])
y = np.array([15,25,35,45])

np.corrcoef(x,y)
```

```python
x = np.array([10,20,30,40])
y = np.array([20,18,15,10])

np.corrcoef(x,y)
```

```python
df = pd.DataFrame({'X': x, 'Y': y})
df.corr()
```

# Skewess
- Skewess measures asymmetry of data
- Shows whether data leans left or right

Types of skewness
1. Zero Skewness: symmetric
2. Positive skewness: Long right tail
3. Negative skewness: Long left tail

## Mathematical Idea of Skewness
$$ Skewness = \frac{E[(x-\mu)^{3}]}{\sigma^3} $$

- X: random variable(data values)
- $\mu$: mean
- $\sigma$: standard deviation

```python
import numpy as np
import pandas as pd

x = np.array([10,20,30,400])
y = np.array([15,25,35,45])

df = pd.DataFrame({'X': x, 'Y': y})

print("Skewness of X column ", df['X'].skew())
print("Skewness of y column ", df['Y'].skew())
```

# Kurtosis
- Kurtosis measures tailedness
- Shows presence of extreme values

Types of Kurtosis:
1. Mesokurtic: normal distribution
2. Leptokurtic: heavy tails, many outliers
3. Platykurtic: light tails

## Mathematical Idea
$$ Skewness = \frac{E[(x-\mu)^{4}]}{\sigma^4} $$

- Uses fourth central moment
- Strongly affected by outliers


```python
x = np.array([10, 20, 30, 400])
df = pd.DataFrame({'X': x})

print('Kurtosis of X column ', df['X'].kurt())
```
