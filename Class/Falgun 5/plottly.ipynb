```python
import plotly.express as px
import pandas as pd
```

```python
# use iris
df = px.data.iris()

# scatter plot
fig = px.scatter(df, x="sepal_length", y="sepal_width", color="species")

fig.show()
fig.write_html("figure.html")
```

```python
figure = px.histogram(df, x="sepal_length", y="species")
figure.show()
```
