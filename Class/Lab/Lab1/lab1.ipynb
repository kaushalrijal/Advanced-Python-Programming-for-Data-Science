```python
import matplotlib.pyplot as plt

class DataSample:
    def __init__(self, name, values):
        self.name = name
        self.values = values

    def summary(self):
        return {
            "mean": sum(self.values)/len(self.values),
            "max": max(self.values),
            "min": min(self.values),
            "range": max(self.values)-min(self.values),
        }
    
    def mean(self, lst):
        return sum(lst)/len(lst)

    def max_value(self, lst):
        return max(lst)

    def min_value(self, lst):
        return min(lst)

    def data_range(self, lst):
        return max(lst) - min(lst)

    def variance(self, lst):
        mean_val = sum(lst)/len(lst)
        return sum((x-mean_val)**2 for x in lst)/len(lst)

    def std_dev(self, lst):
        return self.variance(lst)**0.5
    
    def display(self):
        print(f"Data Sample: {self.name}")
        print("Values:", self.values)

    def box_plot(self):
        plt.figure(figsize=(10, 6))
        plt.boxplot(self.values)
        plt.title(f"Box plot of {self.name}")
        plt.show()

    def frequency(self):
        aDict = {}

        for val in self.values:
            if not val in aDict.keys():
                aDict[val] = 1
            else:
                aDict[val] += 1

        return aDict

    def bar_plot(self):
        frequency = self.frequency()
        plt.figure(figsize=(10, 6))
        plt.bar(frequency.keys(), frequency.values(), width=0.1)
        plt.title(f"Bar Plot of {self.name}")
        plt.xlabel("Values")
        plt.ylabel("Frequencies")
        plt.show()
    
        
```

```python
sample2 = DataSample("Sample2", [10, 20, 30, 40, 10])
sample2.box_plot()
```

```python
sample1 = DataSample("Sample1", [1,2,3,4,4,3,2,3,3,1,3,4, 2, 4])
sample1.bar_plot()
```

```python
# 18. Implement a class Employee with attributes name and salary and a method to apply a 10% raise
class Employee:
    def __init__(self, name, salary):
        self.name = name
        self.salary = salary
    
    def raise_salary(self):
        self.salary += self.salary/10

    def show(self):
        print("Name:", self.name)
        print("Salary:", self.salary)

e1 = Employee("Manish", 100000)
e1.show()
e1.raise_salary()
e1.show()
```
