```python
import csv

data = [
    ["Name", "Age", "City"],
    ["Kaushal", 20, "Biratnagar"],
    ["Guru", 22, "Mirchaiya"]
]

with open("output.csv", "w") as f:
    writer = csv.writer(f)

    writer.writerows(data)
```

```python
import csv

data = [
    {"Name": "Kaushal", "Age": 20},
    {"Name": "Guru", "Age": 22}
]

with open("dict_output.csv", "w") as f:
    writer = csv.DictWriter(f, fieldnames=["Name", "Age"])

    writer.writeheader()
    writer.writerows(data)
```
