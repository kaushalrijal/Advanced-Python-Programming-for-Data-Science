```python
# 1. Write a python program to check if a given numbeer is prime.
import math

def is_prime(num: int) -> bool:
    flag = False
    for i in range(2, int(math.sqrt(num)) + 1):
        if num%i == 0:
            flag = True
            break
    
    return not flag

is_prime(11)
```

```python
# 2. Create a function that converts a list of temperatures from Celsius to Farenheit.

def farenheit(temps:list[float]) -> list[float]:
    faren = []  
    for temp in temps:
        far_value = temp * 1.8 + 32
        faren.append(far_value)
    
    return faren

celsius_temps = [0, 36, 100, 14]

faren_temps = farenheit(celsius_temps)
faren_temps
```

```python
# 3. Write a program to count the number of vowels in a given string.

def num_of_vowels(string:str) -> int:
    vowel_count = 0
    vowels = ['a', 'e', 'i', 'o', 'u']
    for char in string:
        if char in vowels:
            vowel_count += 1

    return vowel_count

num_of_vowels("bananarama")
```

```python
# 4. Implement a function to find the factorial of a number using recursion.

def fact(n:int)->int:
    if(n==0 or n==1):
        return 1
    return n * fact(n-1)

fact(6)
```

```python
# 5. Write a Python script to merge two dictionaries into one

def merge_dict(dict1:dict, dict2:dict)->dict:
    new_dict = {}

    for key in dict1.keys():
        new_dict[key] = dict1[key]

    for key in dict2.keys():
        new_dict[key] = dict2[key]

    return new_dict

dict1 = {
    "a": 1,
    "b": 2
}
dict2 = {
    "c": 3,
    "d": 4
}

dict_merged = merge_dict(dict1, dict2)
dict_merged
```

```python
# 6. Create a functin that takes a list of numbers and returns only the even numbers

def evens(num_list:list[int])->list[int]:
    return list(filter(lambda x: x%2==0, num_list))

evens([1, 2, 3, 4, 5, 6])
```

```python
# 7. Write a program to check if a string is a palindrome.

def is_palindrome(string:str)->bool:
    for i in range(0, int(len(string)/2)):
        if string[i] != string[-(i+1)]:
            return False
    return True

is_palindrome("papa")
```

```python
# 8. Implement a class Cricle with a method to calculate the area and circumference.

class Circle:
    PI = 3.141592654

    def __init__(self, radius):
        self.radius = radius    

    def area(self):
        return self.PI * (self.radius **2)

    def circumference(self):
        return 2 * self.PI * self.radius

circle = Circle(30)
print(circle.area(), circle.circumference())
```

```python
# 9. Write a Python function that reads a text file and counts the number of lines.

def num_of_lines(filename:str)->int:
    lines = 0
    with open(filename, mode="r") as f:
        lines_array = f.readlines()
        lines = len(lines_array)
    
    return lines

num_of_lines("test.txt")
```

```python
# 10. Create a program to sort a list of tuples based on the second element

def sort_tuples(tuples_list: list[tuple])->list[tuple]:
    return sorted(tuples_list, key=lambda x: x[1])

tuples = [(1, "Ram"), (2, "Shyam"), (3, "Hari"), (4, "Basanta")]
sort_tuples(tuples)
```

```python
# 11. Implement a function to remove duplciates from a list without uisng set().

def remove_duplicates(dup: list)->list:
    removed = []

    for item in dup:
        if not item in removed:
            removed.append(item)

    return removed

remove_duplicates([1, 2, 1, 1, 3, 2, 5, 7, 3])
```

```python
# 13. Create a class Rectangel with methods to calculate area and perimeter

class Rectangle:
    def __init__(self, length, width):
        self.length = length
        self.width = width
    
    def area(self):
        return self.length * self.width
    
    def perimeter(self):
        return 2 * (self.length + self.width)
    
rect = Rectangle(5, 3)
rect.area(), rect.perimeter()
```

```python
# 14. Write a function to reverse a list without using the built-in reverse() method.

def reverse(array: list)->list:
    reversed_list = []

    for i in range(len(array)):
        reversed_list.append(array[-(i+1)])
    
    return reversed_list

reverse([1, 2, 5, 3, 4])
```

```python
# 15. Implement a Python function that generates the Fibonacci sequence up to the n terms

def fibonacci(n: int)->list[int]:
    fibo = [0, 1]

    for i in range(2, n):
        fibo.append(fibo[-1] + fibo[-2])
    
    return fibo

fibonacci(9)
```

```python
# 16. Write a progra to count the frequency of each word in a given string

def frequency(sentence:str)->dict:
    freq_dict={}
    words_array = sentence.split()
    for word in words_array:
        if word in freq_dict.keys():
            freq_dict[word] += 1
        else: 
            freq_dict[word] = 1

    return freq_dict

frequency("A big blue fox jumps over another big blue lazy dog over the big pond")
```

```python
# 18. Implement a class Employee with attributes name and salary and a method to apply a 10% raise.
class Employee:
    def __init__(self, name, salary):
        self.name = name
        self.salary = salary
    
    def raise_salary(self):
        self.salary += self.salary/10

    def show(self):
        print("Name:", self.name)
        print("Salary:", self.salary)

e1 = Employee("Manish", 100000)
e1.show()
e1.raise_salary()
e1.show()
```

```python
# 20. Create a function that calculates the sum of digits of a given number

def sum_of_digits(num:int)->int:
    casted_num = str(num)
    sum = 0

    for i in casted_num:
        sum += int(i)

    return sum

sum_of_digits(91)
```
