```python
import pandas as pd

data = {
    "Name": ["Ram", "Shyam", "Bishnu", "Brahma", "Krishna", "Laxmi"],
    "Salary": [1000, 1050, 2000, 2010, 1090, 999]
}

df = pd.DataFrame(data)
print(df.info())
```

```python
df["Gender"] = df["Name"].apply(lambda x:"Female" if x == "Laxmi" else "Male")
df.tail()
```

```python
df.plot.bar(x = "Name", y="Salary", title="Name vs salary")
```

```python
df_A = pd.DataFrame({
    'Roll': [1, 2, 3],
    'Name': ['Ram', 'Shyam', 'Hari']
})
df_B = pd.DataFrame({
    'Roll': [1, 2, 4],
    'Percentage': [60.0, 70.0, 80.0]
})

print("INNER MERGE:")
inner_merge = pd.merge(df_A, df_B, on="Roll", how="inner")
print(inner_merge)
```

```python

print("OUTER MERGE:")
outer_merge = pd.merge(df_A, df_B, on="Roll", how="outer")
print(outer_merge)
```

```python

print("LEFT MERGE:")
left_merge = pd.merge(df_A, df_B, on="Roll", how="left")
print(left_merge)
```

```python

print("RIGHT MERGE:")
right_merge = pd.merge(df_A, df_B, on="Roll", how="right")
print(right_merge)
```
