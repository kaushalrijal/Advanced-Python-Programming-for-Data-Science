# CSV(Comma Seperated Values)

```python
#using the csv module
import csv

with open(file="try.csv", mode="r") as f:
    reader = csv.reader(f)
    for row in reader:
        print(row)
```

```python
#using the dictionary writer
import csv

with open(file="try.csv", mode="r") as f:
    reader = csv.DictReader(f)
    for row in reader:
        print(row)
```

```python

# writing using writer
data = [
    ['Name', 'Address', 'Age'],
    ['Ram', 'Biratnagar', 22],
    ['Shyam', 'Itahari', 23],
    ['Hari', 'Dharan', 21]
]

with open(file="newnew.csv", mode="w", newline="") as f:
    writer = csv.writer(f)
    writer.writerows(data)
```

```python
import csv
data = [
    {'Name': 'Ram', 'Address': 'Biratnagar', 'Age': 22},
    {'Name': 'Shyam', 'Address': 'Itahari', 'Age': 23},
    {'Name': 'Hari', 'Address': 'Dharan', 'Age': 21},
]

with open(file="newnewnew.csv", mode="w", newline=""):
    dictionaryWriter = csv.DictWriter(f, fieldnames=["Name", "Address", "Age"])
    dictionaryWriter.writeheader()
    dictionaryWriter.writerows(data)
```
