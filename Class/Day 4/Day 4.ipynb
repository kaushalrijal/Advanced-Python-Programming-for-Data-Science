```python
# Simple Inheritance
# Base Class
class Animal:
    def __init__(self, name):
        self.name = name

    def speak(self):
        print(f"{self.name} is Speaking")

# Derived Class
class Dog(Animal):
    def __init__(self, name, breed):
        super().__init__(name)
        self.breed = breed

    def speak(self):
        print(f"{self.breed} {self.name} is Barking!")


dogone = Dog("Lucy", "German Shepherd")
dogone.speak()
```

```python
# Multiple Inheritance
class A:
    def show(self):
        print("A")

class B:
    def show(self):
        print("B")

class C(A, B):
    pass

c = C()
c.show()
```

```python
# Method Resolution Order
C.mro()
```

```python
# make a class called DataList, write a program that will compute the mean of the list
class DataList:
    def __init__(self, alist):
        self.alist = alist
    
    def mean(self):
        return sum(self.alist)/len(self.alist)

l1 = DataList([1, 2, 3, 4, 5])
l1.mean()
```
