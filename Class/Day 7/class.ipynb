```python
import requests
url = "https://ekantipur.com/"
response = requests.get(url)
if response.status_code == 200:
    data = response.text
    soup = BeautifulSoup(data, "html.parser")
    print(soup.prettify())
else:
    print(f"Response {response.status_code}")
```

```python
headlines = soup.find_all("h2")
for headline in headlines:
    print(headline.text)
```

```python
import pandas as pd

news = soup.find_all("h2")
final_data = []

for val in news:
    text_data = val.text
    link_data = val.a
    
    if link_data:
        link = link_data.get("href")
        # link = link_data["href"]
    if len(text_data) > 20:
        final_data.append({
            "headline": text_data, 
            "link": link
        })

df = pd.DataFrame(final_data)
df.to_csv("news.csv", index=False)
```
